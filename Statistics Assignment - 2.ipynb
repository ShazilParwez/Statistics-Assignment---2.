````python
# 1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
# After generating the list, find the following:



#   (i) Write a Python function to calculate the mean of a given list of numbers.

# Create a function to find the median of a list of numbers.



#   (ii) Develop a program to compute the mode of a list of integers.



#   (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.



#   (iv) Write a Python function to find the geometric mean of a list of positive numbers.



#   (v) Create a program to calculate the harmonic mean of a list of values.



#   (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).



#   (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.


# Answer : 1

import random
from statistics import mean, median, mode, harmonic_mean
from scipy.stats import trim_mean
import math

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean and Median
def calculate_mean(numbers):
    return mean(numbers)

def calculate_median(numbers):
    return median(numbers)

# (ii) Mode
def calculate_mode(numbers):
    try:
        return mode(numbers)
    except:
        return "Mode is not unique"

# (iii) Weighted Mean
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    return weighted_sum / sum(weights)

# (iv) Geometric Mean
def calculate_geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# (v) Harmonic Mean
def calculate_harmonic_mean(numbers):
    return harmonic_mean(numbers)

# (vi) Midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Trimmed Mean
def calculate_trimmed_mean(numbers, proportion_to_cut):
    return trim_mean(numbers, proportion_to_cut)

# Example usage
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
weights = [random.randint(1, 5) for _ in range(100)]  # Example weights for weighted mean
print("Weighted Mean:", calculate_weighted_mean(int_list, weights))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (5%):", calculate_trimmed_mean(int_list, 0.05))
````

````python
#  2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
# After generating the list, find the following:


#   (i) Compare the given list of visualization for the given data:



#     1. Frequency & Gaussian distribution

#     2. Frequency smoothened KDE plot

#     3. Gaussian distribution & smoothened KDE plot


#   (ii) Write a Python function to calculate the range of a given list of numbers.


#   (iii) Create a program to find the variance and standard deviation of a list of numbers.


#   (iv) Implement a function to compute the interquartile range (IQR) of a list of values.


#   (v) Build a program to calculate the coefficient of variation for a dataset.



#   (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.


#   (vii) Create a program to calculate the quartile deviation of a list of values.



#   (viii) Implement a function to find the range-based coefficient of dispersion for a dataset.


# Answer : 2

import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import iqr

# Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualizations: Frequency & Gaussian Distribution, KDE Plots
def plot_visualizations(data):
    plt.figure(figsize=(15, 10))

    # 1. Frequency & Gaussian Distribution
    plt.subplot(3, 1, 1)
    sns.histplot(data, kde=False, bins=30, color="skyblue", label="Frequency")
    sns.kdeplot(data, color="red", label="Gaussian KDE")
    plt.title("Frequency & Gaussian Distribution")
    plt.legend()

    # 2. Frequency Smoothened KDE Plot
    plt.subplot(3, 1, 2)
    sns.histplot(data, kde=True, bins=30, color="skyblue")
    plt.title("Frequency Smoothened KDE Plot")

    # 3. Gaussian Distribution & Smoothened KDE Plot
    plt.subplot(3, 1, 3)
    sns.kdeplot(data, color="green", label="Gaussian KDE")
    sns.kdeplot(data, bw_adjust=0.5, color="purple", label="Smoothened KDE")
    plt.title("Gaussian Distribution & Smoothened KDE Plot")
    plt.legend()

    plt.tight_layout()
    plt.show()

# (ii) Range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Variance and Standard Deviation
def calculate_variance(numbers):
    return np.var(numbers)

def calculate_standard_deviation(numbers):
    return np.std(numbers)

# (iv) Interquartile Range (IQR)
def calculate_iqr(numbers):
    return iqr(numbers)

# (v) Coefficient of Variation
def calculate_coefficient_of_variation(numbers):
    return np.std(numbers) / np.mean(numbers)

# (vi) Mean Absolute Deviation (MAD)
def calculate_mad(numbers):
    mean_val = np.mean(numbers)
    return np.mean([abs(x - mean_val) for x in numbers])

# (vii) Quartile Deviation
def calculate_quartile_deviation(numbers):
    q1, q3 = np.percentile(numbers, [25, 75])
    return (q3 - q1) / 2

# (viii) Range-based Coefficient of Dispersion
def calculate_range_based_coefficient_of_dispersion(numbers):
    return (max(numbers) - min(numbers)) / (max(numbers) + min(numbers))

# Example usage
print("Range:", calculate_range(int_list2))
print("Variance:", calculate_variance(int_list2))
print("Standard Deviation:", calculate_standard_deviation(int_list2))
print("Interquartile Range (IQR):", calculate_iqr(int_list2))
print("Coefficient of Variation:", calculate_coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mad(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", calculate_range_based_coefficient_of_dispersion(int_list2))

# Display visualizations
plot_visualizations(int_list2)
````

````python
# 3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

# Answer : 3

class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initializes the DiscreteRandomVariable with outcomes and their associated probabilities.

        Parameters:
        outcomes (list): A list of possible outcomes (values) of the random variable.
        probabilities (list): A list of probabilities associated with each outcome.
                              The length of probabilities should match the length of outcomes.
                              The sum of all probabilities should be 1.
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("Outcomes and probabilities must have the same length.")
        if not abs(sum(probabilities) - 1) < 1e-6:
            raise ValueError("The sum of probabilities must be 1.")

        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculates the expected value (mean) of the discrete random variable.

        Returns:
        float: The expected value.
        """
        return sum(outcome * prob for outcome, prob in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculates the variance of the discrete random variable.

        Returns:
        float: The variance.
        """
        mean = self.expected_value()
        return sum(prob * ((outcome - mean) ** 2) for outcome, prob in zip(self.outcomes, self.probabilities))

# Example:
# Define outcomes and their corresponding probabilities
outcomes = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create an instance of the DiscreteRandomVariable class
random_variable = DiscreteRandomVariable(outcomes, probabilities)

# Calculate the expected value and variance
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

````python
# 4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

# Answer : 4

import random

class DieSimulation:
    def __init__(self, sides=6, num_rolls=1000):
        """
        Initializes the DieSimulation with the number of sides on the die and the number of rolls.

        Parameters:
        sides (int): The number of sides on the die (default is 6 for a fair six-sided die).
        num_rolls (int): The number of times to roll the die (default is 1000).
        """
        self.sides = sides
        self.num_rolls = num_rolls
        self.outcomes = []

    def roll_die(self):
        """
        Simulates rolling the die num_rolls times and stores the outcomes.
        """
        self.outcomes = [random.randint(1, self.sides) for _ in range(self.num_rolls)]

    def calculate_expected_value(self):
        """
        Calculates the expected value (mean) of the die roll outcomes.

        Returns:
        float: The expected value of the outcomes.
        """
        return sum(self.outcomes) / len(self.outcomes)

    def calculate_variance(self):
        """
        Calculates the variance of the die roll outcomes.

        Returns:
        float: The variance of the outcomes.
        """
        mean = self.calculate_expected_value()
        return sum((x - mean) ** 2 for x in self.outcomes) / len(self.outcomes)


# Example usage
die_sim = DieSimulation(sides=6, num_rolls=1000)
die_sim.roll_die()

# Calculate and display the expected value and variance
expected_value = die_sim.calculate_expected_value()
variance = die_sim.calculate_variance()

print("Expected Value (Simulated):", expected_value)
print("Variance (Simulated):", variance)
````

````python
#5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

#Answer : 5

import numpy as np

def sample_distribution(dist_type, params, sample_size=1000):
    """
    Generates random samples from a specified distribution and calculates the mean and variance.

    Parameters:
    dist_type (str): The type of distribution ("binomial", "poisson", etc.).
    params (dict): Parameters required for the specified distribution.
                   - For "binomial": {"n": number of trials, "p": probability of success}
                   - For "poisson": {"lambda": average rate of occurrence}
    sample_size (int): The number of random samples to generate (default is 1000).

    Returns:
    tuple: A tuple containing (samples, mean, variance).
    """
    if dist_type == "binomial":
        n = params.get("n")
        p = params.get("p")
        samples = np.random.binomial(n, p, sample_size)

    elif dist_type == "poisson":
        lam = params.get("lambda")
        samples = np.random.poisson(lam, sample_size)

    else:
        raise ValueError("Unsupported distribution type. Choose 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return samples, mean, variance

# Example usage for a binomial distribution
binomial_samples, binomial_mean, binomial_variance = sample_distribution(
    "binomial", params={"n": 10, "p": 0.5}, sample_size=1000
)
print("Binomial Distribution - Mean:", binomial_mean)
print("Binomial Distribution - Variance:", binomial_variance)

# Example usage for a Poisson distribution
poisson_samples, poisson_mean, poisson_variance = sample_distribution(
    "poisson", params={"lambda": 5}, sample_size=1000
)
print("Poisson Distribution - Mean:", poisson_mean)
print("Poisson Distribution - Variance:", poisson_variance)
````

````python
#7. Use seaborn library to load tips dataset. Find the following from the dataset for the columns total_bill and tip`:



#   (i) Write a Python function that calculates their skewness.


#   (ii) Create a program that determines whether the columns exhibit positive skewness, negative skewness, or is
# approximately symmetric.


#   (iii) Write a function that calculates the covariance between two columns.


#   (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.


#   (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
# scatter plots


# Answer : 6

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load the tips dataset from seaborn
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness of a column
def calculate_skewness(column):
    return skew(column, bias=False)

# (ii) Program to determine skewness type
def skewness_type(column):
    skew_value = calculate_skewness(column)
    if skew_value > 0:
        return "Positive Skewness"
    elif skew_value < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# (iii) Function to calculate covariance between two columns
def calculate_covariance(column1, column2):
    return np.cov(column1, column2, bias=True)[0][1]

# (iv) Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(column1, column2):
    return np.corrcoef(column1, column2)[0][1]

# (v) Visualization using scatter plot
def visualize_correlation(df, col1, col2):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=col1, y=col2)
    plt.title(f'Scatter Plot: {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.grid(True)
    plt.show()

# Applying functions to 'total_bill' and 'tip' columns
total_bill_skewness = calculate_skewness(tips['total_bill'])
tip_skewness = calculate_skewness(tips['tip'])

print("Skewness:")
print(f"Total Bill: {total_bill_skewness} ({skewness_type(tips['total_bill'])})")
print(f"Tip: {tip_skewness} ({skewness_type(tips['tip'])})\n")

covariance = calculate_covariance(tips['total_bill'], tips['tip'])
print(f"Covariance between Total Bill and Tip: {covariance}\n")

pearson_corr = calculate_pearson_correlation(tips['total_bill'], tips['tip'])
print(f"Pearson Correlation Coefficient between Total Bill and Tip: {pearson_corr}\n")

# Visualizing correlation
visualize_correlation(tips, 'total_bill', 'tip')
````

````python
#8. Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

# Answer : 8

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    x (float): The value at which to evaluate the PDF.
    mean (float): The mean () of the normal distribution.
    std_dev (float): The standard deviation () of the normal distribution.

    Returns:
    float: The PDF value for the given x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    return coefficient * exponent

# Example usage
mean = 0
std_dev = 1
x_value = 1.5

pdf_value = normal_pdf(x_value, mean, std_dev)
print(f"The PDF value at x = {x_value} for N({mean}, {std_dev}^2) is: {pdf_value}")
````

````python
#9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

#Answer : 9

import math

def exponential_cdf(x, rate_parameter):
    """
    Calculate the CDF of an exponential distribution.

    Parameters:
    x (float): The value at which to evaluate the CDF.
    rate_parameter (float): The rate parameter () of the distribution ( > 0).

    Returns:
    float: The CDF value for the given x.
    """
    if x < 0:
        return 0.0
    return 1 - math.exp(-rate_parameter * x)

# Example usage
rate_parameter = 1.5  #  (rate parameter)
x_value = 2.0  # Point at which to evaluate the CDF

cdf_value = exponential_cdf(x_value, rate_parameter)
print(f"The CDF value at x = {x_value} for an exponential distribution with  = {rate_parameter} is: {cdf_value}")
````

````python
#10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

#Answer : 10

import math

def poisson_pmf(k, rate_parameter):
    """
    Calculate the PMF of a Poisson distribution.

    Parameters:
    k (int): The number of occurrences (must be a non-negative integer).
    rate_parameter (float): The rate parameter () of the Poisson distribution ( > 0).

    Returns:
    float: The PMF value for the given k.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    if rate_parameter <= 0:
        raise ValueError("Rate parameter () must be positive.")

    return (math.exp(-rate_parameter) * (rate_parameter ** k)) / math.factorial(k)

# Example usage
rate_parameter = 3.0  #  (average rate of occurrences)
k_value = 5  # Number of occurrences

pmf_value = poisson_pmf(k_value, rate_parameter)
print(f"The PMF value for k = {k_value} and  = {rate_parameter} is: {pmf_value}")
````

````python
# 11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.


# To generate the data use the following command:

# ```python

# import numpy as np

# # 50 purchases out of 1000 visitors

# old_layout = np.array([1] * 50 + [0] * 950)

# # 70 purchases out of 1000 visitors

# new_layout = np.array([1] * 70 + [0] * 930)

#   ```

# Apply z-test to find which layout is successful.


#Answer : 11

import numpy as np
from scipy.stats import norm

# Generate data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Sample sizes
n_old = len(old_layout)
n_new = len(new_layout)

# Conversion counts
x_old = sum(old_layout)
x_new = sum(new_layout)

# Proportions
p_old = x_old / n_old
p_new = x_new / n_new

# Pooled proportion
p_pooled = (x_old + x_new) / (n_old + n_new)

# Z-test statistic
z = (p_new - p_old) / np.sqrt(p_pooled * (1 - p_pooled) * (1/n_old + 1/n_new))

# Calculate p-value
p_value = 1 - norm.cdf(z)

# Output results
print(f"Old Layout Conversion Rate: {p_old}")
print(f"New Layout Conversion Rate: {p_new}")
print(f"Z-Statistic: {z}")
print(f"P-Value: {p_value}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

````python
# 12. A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, and their scores before and after the program were recorded.


# Use the below code to generate samples of respective arrays of marks:

# ```python

# before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

# after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# ```

# Use z-test to find if the claims made by tutor are true or false.


#Answer 12:

import numpy as np
from scipy.stats import norm

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Differences
differences = after_program - before_program

# Mean and standard deviation of differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(differences)

# Z-test statistic
z = mean_diff / (std_diff / np.sqrt(n))

# Calculate p-value (one-tailed test)
p_value = 1 - norm.cdf(z)

# Output results
print(f"Mean Difference: {mean_diff}")
print(f"Standard Deviation of Differences: {std_diff}")
print(f"Z-Statistic: {z}")
print(f"P-Value: {p_value}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The tutoring program improves scores.")
else:
    print("Fail to reject the null hypothesis: No significant improvement in scores.")
````

````python
# 21. A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.

# Use the below code to generate data:

# ```python

# np.random.seed(0)

# age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

# voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)


#Answer : 21

import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a contingency table
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Chi-Square test calculation
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2 Statistic: {chi2:.2f}, P-value: {p_value:.4f}, Degrees of Freedom: {dof}")
````

````python
# 22. A company conducted a customer satisfaction survey to determine if there is a significant relationship between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a ChiSquare test to determine if there is a significant relationship between product satisfaction levels and customer regions.

# Sample data:

# ```python

# #Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)

# data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])


#Answer : 22

import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

# Chi-Square test calculation
chi2, p_value, dof, expected = chi2_contingency(data)

print(f"Chi2 Statistic: {chi2:.2f}, P-value: {p_value:.4f}, Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)
````

````python
# 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores.

#   Use the following data:

#   ```python

#   # Sample data: Customer satisfaction scores for each product version

#   standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

#   premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

#   deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]


#Answer : 24

from scipy.stats import f_oneway

# Sample data
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# ANOVA test calculation
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

print(f"F-statistic: {f_stat:.2f}, P-value: {p_value:.4f}")
````
